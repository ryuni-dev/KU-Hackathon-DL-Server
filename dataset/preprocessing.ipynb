```python
data = pd.DataFrame(columns=['Sentiment', 'Sentence'])
```

```python
df = df[['감정_대분류','사람문장1','사람문장2','사람문장3','사람문장4']]
df.head()
```

```python
df['감정_대분류'].unique()
```

```python
def sentiment2idx(sentiment):
    if sentiment=='기쁨':
        return 0
    elif sentiment=='기쁨 ':
        return 0
    elif sentiment=='불안':
        return 1
    elif sentiment=='불안 ':
        return 1
    elif sentiment=='당황':
        return 2
    elif sentiment=='슬픔':
        return 3
    elif sentiment=='분노':
        return 4
    elif sentiment=='상처':
        return 5
```

```python
idx = 0
for i in range(len(df)):
    sentiment = df.iloc[i,0]
    sentiment = sentiment2idx(sentiment)
    for j in range(1,5):
        sentence = df.iloc[i, j]
        data.loc[idx] = [sentiment, sentence]
        idx += 1
```

```python
data['Sentiment'] = data['Sentiment'].astype(int) 
```
